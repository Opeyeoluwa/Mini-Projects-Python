```python
def fibonacci(n):
    if n<0:
        print('Incorrect Input')
    elif n==1:
        return 0
    elif n==2:
        return 1
    else:
        return fibonacci(n-1)+ fibonacci(n-2)
```

```python
n= int(input('How many terms would you like to generate? '))
def fibonacci(n):
    n1=0
    n2=1
    count=0
    if n<=0:
        print('Please enter a positive Integer')
    elif n==1:
        print('Fibonacci sequence up to',n,':')
        print(n1)
    else:
        print('Fibonacci sequence up to',n,':')
        while count < n:
            print(n1, end=',')
            nth= n1+n2
            n1=n2
            n2=nth
            count +=1
fibonacci(n)
```

```python
def fibonacci2(n):
    if n<=1:
        return n
    else:
        return(fibonacci2(n-1) + fibonacci2(n-2))
```

```python
nterms= int(input('How many terms? '))
if nterms<=0:
    print('Please Enter a Postive Integer')
else:
    print('Fibonacci Sequence:')
    for i in range(nterms):
        print(fibonacci2)
```

```python
import pygame, sys
from pygame.locals import *
from pygame import *

pygame.init()
display_width = 800
display_height = 600

blue=( 0, 0, 255, 100)
fuchsia=(255, 0, 255, 100)
gray=(128, 128, 128, 100) 
black = (0,0,0)
white = (255,255,255)
red = (200,0,0)
green = (0, 200, 0)
light_red = (255,0,0)
light_green = (0,255,0)
block_color = (53,115,255)
 
gameDisplay = pygame.display.set_mode((display_width,display_height))
pygame.display.set_caption('Dead & Injured')
clock = pygame.time.Clock()

pixAr = pygame.PixelArray(gameDisplay) #the array portion, to enable easy drawing

def text_objects(text, font): #function for the home page text
    textSurface = font.render(text, True, black)
    return textSurface, textSurface.get_rect()

def button_text(text, font): #function for the button color
    textsurface = font.render(text, True, white)
    return textsurface, textsurface.get_rect()

def main():
    font = pygame.font.Font(None, 32)
    input_box = pygame.Rect(100, 100, 140, 32)
    color_inactive = pygame.Color('lightskyblue3')
    color_active = pygame.Color('dodgerblue2')
    color = color_inactive
    active = False
    text = ''
    done = False

    while not done:
        for event in pygame.event.get():
            print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
            if event.type == pygame.MOUSEBUTTONDOWN:
                # If the user clicked on the input_box rect.
                if input_box.collidepoint(event.pos):
                    # Toggle the active variable.
                    active = not active
                else:
                    active = False
                # Change the current color of the input box.
                color = color_active if active else color_inactive
            if event.type == pygame.KEYDOWN:
                if active:
                    if event.key == pygame.K_RETURN:
                        print(text)
                        text = ''
                    elif event.key == pygame.K_BACKSPACE:
                        text = text[:-1]
                    else:
                        text += event.unicode
        # Render the current text.
        txt_surface = font.render(text, True, color)
        # Resize the box if the text is too long.
        width = max(200, txt_surface.get_width()+10)
        input_box.w = width
        # Blit the text.
        gameDisplay.blit(txt_surface, (input_box.x+5, input_box.y+5))
        # Blit the input_box rect.
        pygame.draw.rect(gameDisplay, color, input_box, 2)

        pygame.display.flip()
        clock.tick(30)

def quitgame():
    pygame.quit()
    quit()

def unpause():
    global pause
    pause = False
    
def button(msg,x,y,w,h,ic,ac, action= None): # msg is message, x= y coordinate, y= x coordinate, h= height, ic=inactivecolor, ac=actuve color
    mouse = pygame.mouse.get_pos() # for getting the mouse position i.e. hovering
    click = pygame.mouse.get_pressed() # for clicking on the mouse
    print(click)
    if x+w > mouse[0] > x and y+h > mouse[1] > y:
        pygame.draw.rect(gameDisplay, ac,(x,y,w,h))
        
        if click[0] == 1 and action != None: # the clicking function
            action()         
    else:
        pygame.draw.rect(gameDisplay, ic,(x,y,w,h))

    smallText = pygame.font.Font("freesansbold.ttf",20)
    textSurf, textRect = button_text(msg, smallText)
    textRect.center = ( (x+(w/2)), (y+(h/2)) )
    gameDisplay.blit(textSurf, textRect)

def game_intro():  #Home page of the game

    intro = True

    while intro:
        for event in pygame.event.get():
            print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
                
        gameDisplay.fill(white)
        largeText = pygame.font.Font('freesansbold.ttf',100)
        TextSurf, TextRect = text_objects("Dead & Injured", largeText)
        TextRect.center = ((display_width/2),(display_height/2))
        gameDisplay.blit(TextSurf, TextRect)
        
        button("BEGIN",150,450,100,50, green, light_green, option)
        button("QUIT",550,450,100,50, red, light_red, quitgame)
        #mouse = pygame.mouse.get_pos() #mouse module for checking for mouse sensitivity close to object
        #print(mouse)
        #if (150+100 > mouse[0] > 150) and (450+50 > mouse[1] > 450): #adding color change to mouse response button1
            #pygame.draw.rect(gameDisplay, light_green,(150,450,100,50))
        #else:
            #pygame.draw.rect(gameDisplay, green,(150,450,100,50)) #Button1 for playing game

        #if (550+100 > mouse[0] > 550) and (450+50 > mouse[1] > 450): #adding color change to mouse response button2
            #pygame.draw.rect(gameDisplay, light_red, (550,450,100,50))
        #else:
            #pygame.draw.rect(gameDisplay, red,(550,450,100,50)) #Button2 for playing game

        #smallText = pygame.font.Font("freesansbold.ttf",20) #Adding text to the buttons
        #textSurf, textRect = button_text("BEGIN", smallText)
        #textRect.center = ( (150+(100/2)), (450+(50/2)) )
        #gameDisplay.blit(textSurf, textRect)

        #smallText = pygame.font.Font("freesansbold.ttf",20) #Adding text to the buttons
        #textSurf, textRect = button_text("QUIT", smallText)
        #textRect.center = ( (550+(100/2)), (450+(50/2)) )
        #gameDisplay.blit(textSurf, textRect)

        pygame.display.update()
        clock.tick(15)

def option():
    hello = True

    while hello:
        for event in pygame.event.get():
            print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
                
        gameDisplay.fill(white)
        largeText = pygame.font.Font('freesansbold.ttf',50)
        TextSurf, TextRect = text_objects("CHOOSE DIFFICULTY", largeText)
        TextRect.center = ((display_width/2),(display_height/4))
        gameDisplay.blit(TextSurf, TextRect)

        button("BEGINNER",335,300,130,50, green, light_red, beginner)
        button("INTERMEDIATE",310,380,180,50, green, light_red, intermediate)
        button("EXPERT",350,460,100,50, green, light_red, expert)

        pygame.display.update()
        clock.tick(15)

def beginner():
    hello = True

    while hello:
        for event in pygame.event.get():
            print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()

        gameDisplay.fill(white)
        
        main()
        pygame.display.update()
        clock.tick(15)

def intermediate():
    hello = True

    while hello:
        for event in pygame.event.get():
            print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
                
        gameDisplay.fill(white)
        main()

        pygame.display.update()
        clock.tick(15)

def expert():
    hello = True

    while hello:
        for event in pygame.event.get():
            print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
                
        gameDisplay.fill(white)
        main()

        pygame.display.update()
        clock.tick(15)

def paused():

    largeText = pygame.font.SysFont("comicsansms",115)
    TextSurf, TextRect = button_text("Game Paused", largeText)
    TextRect.center = ((display_width/2),(display_height/2))
    gameDisplay.blit(TextSurf, TextRect)
    

    while pause:
        for event in pygame.event.get():
            #print(event)
            if event.type == pygame.QUIT:
                pygame.quit()
                quit()
                
        gameDisplay.fill(white)        
        button("CONTINUE",150,450,100,50,green,bright_green,unpause)
        button("QUIT",550,450,100,50,red,bright_red,quitgame)

        pygame.display.update()
        clock.tick(15)   
del pixAr
game_intro()
```
